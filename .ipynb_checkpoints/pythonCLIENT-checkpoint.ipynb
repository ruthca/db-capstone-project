```python
query = """
SELECT 
    deccustomers.CustomerName,
    deccustomers.CustomerPhone,
    decorders.TotalCost
FROM deccustomers
JOIN decorders
    ON deccustomers.idCustomer = decorders.idCustomer
WHERE decorders.TotalCost > 60;
"""

cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(row)
```
